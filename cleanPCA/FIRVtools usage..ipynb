```python

x = ["EUR", "GBP", "JPY", "USD"]
fwd, par = firv.loadCurrencies(x)
```

```python
start = dt.datetime(2014, 1, 1)
end = dt.datetime(2016,1, 1)
print (fwd["USD"])
target = ["2y", "5y", "7y", "10y", "20y", "30y"]
comp = ["2y", "2s10s", "2s5s10s"]

usdFwdplot, usdPCA = firv.plotSummary("USD", "fwd", start, end, parD=par, fwdD=fwd, 
                                      tenors=target, AR=2, resample=1, cTrades=comp)

usdFwdplot.savefig("test2.png")
plt.show()
```
